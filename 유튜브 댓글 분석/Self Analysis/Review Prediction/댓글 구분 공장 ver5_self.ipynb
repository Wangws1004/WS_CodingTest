## 긍정,부정 분류 모델 적용

```python
import numpy as np
import tensorflow as tf
from keras.models import load_model
from transformers import BertTokenizer
from keras.preprocessing.sequence import pad_sequences
import pandas as pd
from collections import Counter
```

```python
# TensorFlow GPU 사용 가능 여부 확인
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"Using GPU: {gpus}")
    except RuntimeError as e:
        print(e)
```

```python
model = load_model('./best_model_self.h5')

# BERT 토크나이저 로드
tokenizer = BertTokenizer.from_pretrained("klue/bert-base")

# 감성 예측 결과를 저장할 Counter 객체 초기화
emotion_counter = Counter()

# df["PRO"] 열을 500개씩 분할하여 예측 수행
batch_size = 500

for batch in np.array_split(df['PRO'], np.ceil(len(df) / batch_size)):
    tokenized_texts = [tokenizer.encode(text, add_special_tokens=True, max_length=128, truncation=True) for text in batch]
    data = pad_sequences(tokenized_texts, maxlen=128, padding='post')
    predictions = model.predict(data)
    predicted_emotions = predictions.argmax(axis=1)
    emotion_counter.update(predicted_emotions)

# 감성 예측 결과 출력
print("감성 예측 결과:")
for emotion, count in emotion_counter.items():
    print(f"감성 {emotion}: {count}개")
```

```python
# 감성 예측 함수
def predict_sentiment(text):
    inputs = tokenizer.encode_plus(text, add_special_tokens=True, max_length=512, truncation=True, padding="max_length", return_tensors="tf")
    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]
    predictions = best_model_self.predict([input_ids, attention_mask])
    return np.argmax(predictions, axis=1)[0]  # 또는 모델 출력에 맞게 조정

# 각 행에 대한 감성 분석 수행
predicted_emotions = []

# tqdm을 사용하여 진행 상황을 확인합니다.
for chunk in tqdm(pd.Series(df["PRO"]).array_split(500), desc="Predicting emotions", unit="chunk"):
    chunk_predictions = [predict_sentiment(text) for text in chunk]
    predicted_emotions.extend(chunk_predictions)

# 예측된 감정을 DataFrame에 추가
df["예측 감정_self"] = predicted_emotions
```
